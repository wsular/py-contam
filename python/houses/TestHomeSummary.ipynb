# Summary plots for contam output for WSU test homes

```python
%pylab inline 

import pandas as pd
from socket import gethostname

hostname = gethostname()
if hostname.rfind('gaia')>=0:
    d = '/mnt/data/lima/iaq/houses/outdoors/'
else:
    d = '/Users/vonw/data/iaq/houses/test_homes/'
```

### Define test homes and separate by season

```python
houses = ['h002_summer',
          'h002_winter',
          'h003_summer',
          'h003_winter',
          'h004_summer',
          'h004_winter',
          'h005_summer',
          'h006_summer',
#          'h006_winter',
          'h007_summer',
          'h007_winter',
          'h008_summer',
          'h008_winter',
          'h009_summer',
          'h009_winter',
          'h010_summer',
          'h010_winter']

summer = []
winter = []
for house in houses:
    if house.rfind('summer')>=0:
        summer.append(house)
    else:
        winter.append(house)
```

## Test Home Temperatures

```python
T = pd.concat([pd.read_hdf(d+'houses.hdf', 'h002_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h002_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h003_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h003_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h004_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h004_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h005_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h006_summer/T'),
#               pd.read_hdf(d+'houses.hdf', 'h006_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h007_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h007_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h008_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h008_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h009_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h009_winter/T'),
               pd.read_hdf(d+'houses.hdf', 'h010_summer/T'),
               pd.read_hdf(d+'houses.hdf', 'h010_winter/T')],keys=houses)
```

```python
figure(figsize=(14,8))
T.loc[summer].groupby(level=0).hist(histtype='step', linewidth=2)
xlabel('Temperature (C)')
ylabel('Number of Occurrences')
title('Test Home Temperatures for Summer')
legend(summer, loc='best')
```

```python
figure(figsize=(14,8))
T.loc[winter].groupby(level=0).hist(histtype='step', linewidth=2)
xlabel('Temperature (C)')
ylabel('Number of Occurrences')
title('Test Home Temperatures for Winter')
legend(winter, loc='best')
```
