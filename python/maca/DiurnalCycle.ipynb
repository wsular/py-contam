```python
t = arange(0,48,3)  # 3-hourly time vector
Tmin = 15.
Tmax = 25.
Tave = mean([Tmin, Tmax])
C    = Tmax - Tave
phase = 15.         # phase at which Tmax occurs

figure(figsize=(20,10))
T = Tave + C*cos(2*pi*(t-phase)/24)
th = arange(0,48,0.25)
Th = Tave + C*cos(2*pi*(th-phase)/24)
plot(t,T,'o',th,Th)
xticks(t)
xlabel('Time (hours)')
ylabel('Temperature C')
title('Generation of diurnal cycle using Tmax and Tmin')
grid()
```
